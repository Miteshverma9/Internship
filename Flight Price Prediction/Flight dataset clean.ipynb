```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Flight_price.csv')
data
```

```python
data.info()
```

```python
a = []
for i in data['Price']:
    a.append(i.replace(',',''))
```

```python
b = []
for i in a:
    b.append(i.replace('1 left at',''))
```

```python
c = []
for i in b:
    c.append(i.replace('2 left at',''))
```

```python
d = []
for i in c:
    d.append(i.replace('3 left at',''))
```

```python
e = []
for i in d:
    e.append(i.replace('4 left at',''))
```

```python
data['Price'] = e
```

```python
data['Stops'].replace('non-stop','Non Stop',inplace = True)
```

```python
data['Stops'].replace('1 stop','1 Stop',inplace = True)
```

```python
data['Flight'].replace('VISTARA','Vistara', inplace = True)
data['Flight'].replace('INDIGO','IndiGo', inplace = True)
data['Flight'].replace('AIRASIA INDIA','Air Asia', inplace = True)
data['Flight'].replace('AIR INDIA','Air India', inplace = True)
data['Flight'].replace('SPICEJET','SpiceJet', inplace = True)
data['Flight'].replace('GO FIRST','Go First', inplace = True)
```

```python
z = []
for i in data['Duration']:
    z.append(i.replace('hr','h'))
```

```python
data['Duration'] = z
```

```python
x = []
for i in data['Duration']:
    x.append(i.replace('min','m'))
```

```python
data['Duration'] = x
```

```python
data.drop('Unnamed: 0',axis = 1)
```

```python
data.to_csv('Flight_price_dataset.csv',index = False)
```
